```python
import pandas as pd
import matplotlib.pyplot as plot
```

```python

# Tranformamos el archivo en un data frame 
direccion = "train-titanic.csv"
data = pd.read_csv(direccion)
```

```python
# Gráfica 1
plot.figure(figsize=(6, 4))
data['Survived'].value_counts().plot(kind='bar', color=['red', 'green'])
plot.title('Distribución de Supervivencia')
plot.xlabel('Supervivió (0 = No, 1 = Sí)')
plot.ylabel('Cantidad')
plot.xticks(rotation=0)
plot.show()
```

```python
# Gráfica 2
plot.figure(figsize=(8, 5))
data['Age'].plot(kind='hist', bins=30, color='skyblue', edgecolor='black')
plot.title('Distribución de Edades')
plot.xlabel('Edad')
plot.ylabel('Frecuencia')
plot.show()
```

```python
# Gráfica 7
plot.figure(figsize=(10, 6))
pd.crosstab([data['Pclass'], data['Sex']], data['Survived']).plot(kind='bar', figsize=(10, 6), color=['purple', 'lime'], edgecolor='black')
plot.title('Supervivencia por Clase y Género')
plot.xlabel('Clase y Género')
plot.ylabel('Cantidad')
plot.xticks(rotation=45)
plot.legend(['No sobrevivió', 'Sobrevivió'])
plot.show()
```

```python
# Gráfica 4: Supervivencia por clase
plot.figure(figsize=(8, 5))
pd.crosstab(data['Pclass'], data['Survived']).plot(kind='bar', figsize=(8, 5), color=['blue', 'orange'], edgecolor='black')
plot.title('Supervivencia por Clase')
plot.xlabel('Clase')
plot.ylabel('Cantidad')
plot.xticks(rotation=0)
plot.legend(['No sobrevivió', 'Sobrevivió'])
plot.show()
```

```python
# Cantidad de valores nulos por columna
missing_values = data.isnull().sum()

# Gráfica 1: Valores nulos por columna
plot.figure(figsize=(8, 4))
missing_values.plot(kind='bar', color='red', edgecolor='black')
plot.title('Valores Nulos por Columna')
plot.xlabel('Columnas')
plot.ylabel('Cantidad de Valores Nulos')
plot.xticks(rotation=45)
plot.show()
```
