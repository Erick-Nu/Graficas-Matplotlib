```python
import matplotlib.pyplot as plt
```

```python
# Funcion Plot
# Descripción: Crear gráficos de líneas.

x = [1,2,3,4]
y = [10,20,25,30]
plt.plot(x,y)
plt.show()
```

```python
# Funcion Scatter 
# Descripción: Crear gráficos de dispersión.

x = [5,7,8,7]
y = [8,5,6,7]
plt.scatter(x,y, color = 'green')
plt.show()
```

```python
# Funcion Bar
# Descripción: Crear gráficos de barras.

categories = ['A', 'B', 'C']
values = [3, 7, 8]
plt.bar(categories, values, color='purple')
plt.show()
```

```python
# Funcion Pie
# Descripción: Crear gráficos de pastel.

labels = ['A', 'B', 'C']
sizes = [15, 30, 45]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.show()
```

```python
# Funcion Imshow
# Descripción: Mostrar imágenes.

import numpy as np
img = np.random.random((10, 10))
plt.imshow(img, cmap='viridis')
plt.show()
```

```python
# Funcion Boxplot
# Descripción: Crear gráficos de caja y bigotes.

data = [7, 8, 5, 6, 9, 4, 5]
plt.boxplot(data)
plt.show()
```

```python
# Funcion Stackplot
# Descripción: Crear gráficos de áreas apiladas.

x = [1, 2, 3, 4]
y1 = [1, 2, 3, 4]
y2 = [2, 2, 3, 4]
plt.stackplot(x, y1, y2, labels=['A', 'B'])
plt.legend()
plt.show()
```

```python
# Funcion Errorbar
# Descripción: Crear gráficos con barras de error.Crear gráficos con barras de error.

x = [1, 2, 3]
y = [2, 4, 6]
error = [0.2, 0.4, 0.1]
plt.errorbar(x, y, yerr=error, fmt='o')
plt.show()
```

```python
# Funcion Fill_between
# Descripción: Rellenar el área entre dos curvas.

x = [1, 2, 3, 4]
y1 = [1, 2, 3, 4]
y2 = [0, 1, 2, 3]
plt.fill_between(x, y1, y2, color='skyblue', alpha=0.4)
plt.show()
```

```python
# Funcion Barh
# barh(x, y): Dibuja un diagrama de barras horizontales donde x es una lista
#  con la posición de las barras en el eje Y, e y es una lista con la longitud de las barras en el eje X. 

fig, ax = plt.subplots()
ax.barh([1, 2, 3], [3, 2, 1])
plt.show()
```

```python
# Funciones con la libreria (Seaborn)
import seaborn as sb
```

```python
# Funcion Heatmap
# Descripción: Crear gráficos de calor.

import numpy as np
data = np.random.rand(5, 5)
sb.heatmap(data, annot=True)
plt.show()
```

```python
# Funcion Pairplot
#  Descripción: Visualizar relaciones entre pares de variables.
# Gráfica: [Par de gráficos]

iris = sb.load_dataset('iris')
sb.pairplot(iris, hue='species')
plt.show()
```

```python
# Funcion Boxplot
# Descripción: Crear gráficos de caja.

sb.boxplot(x='species', y='sepal_width', data=iris)
plt.show()
```

```python
# Funcion Violinplot
# Descripción: Crear gráficos de violín.

sb.violinplot(x='species', y='sepal_length', data=iris)
plt.show()
```

```python
# Funcion Barplot
# Descripción: Crear gráficos de barras.

sb.barplot(x='species', y='petal_length', data=iris)
plt.show()
```

```python
# Funciones con la libreria (Pandas and plot)
import pandas as pd
```

```python
# Funcion plot.line
# Descripción: Crear gráficos de líneas.

# Datos en un DataFrame
datos = {"Mes": ["Enero", "Febrero", "Marzo", "Abril"],
    "Ventas": [200, 300, 400, 350]}
df = pd.DataFrame(datos)


df.plot(x="Mes", y="Ventas", kind="line", title="Ventas por Mes", ylabel="Ventas")
```

```python
# Funcion Plot.bar
# Descripción: Crear gráficos de barras.

# Datos en un DataFrame
datos = {"Producto": ["A", "B", "C"],
    "Cantidad": [10, 15, 7]}
df = pd.DataFrame(datos)

# Crear gráfico de barras
df.plot(x="Producto", y="Cantidad", kind="bar", title="Cantidad por Producto")
```

```python
# Funcion Plot.scatter
# Descripción: Crear gráficos de dispersión.

# Datos en un DataFrame
datos = {"Altura": [150, 160, 165, 170, 175],
        "Peso": [50, 55, 60, 65, 70]}
df = pd.DataFrame(datos)

# Crear gráfico de dispersión
df.plot(x="Altura", y="Peso", kind="scatter", title="Altura vs Peso")
```

```python
# Funcion Plot.hist
# Descripción: Crear histogramas.   

# Datos en un DataFrame
datos = {"Edades": [23, 25, 25, 30, 35, 30, 40, 25, 35, 23, 25]}
df = pd.DataFrame(datos)

# Crear histograma
df["Edades"].plot(kind="hist", title="Distribución de Edades")
```

```python
# Funcion Plot.pie
# Descripción: Crear gráficos de pastel.

# Datos en un DataFrame
datos = {"Categoría": ["A", "B", "C", "D"],
        "Porcentaje": [30, 40, 20, 10]}
df = pd.DataFrame(datos)

# Crear gráfico de pastel
df.set_index("Categoría")["Porcentaje"].plot(kind="pie", autopct="%1.1f%%", title="Distribución de Categorías")


```
